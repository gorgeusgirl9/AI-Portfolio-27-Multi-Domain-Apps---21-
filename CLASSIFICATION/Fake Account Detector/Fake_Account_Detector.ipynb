```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import joblib

# 1. Sentetik Sosyal Medya Verisi Oluşturma
np.random.seed(42)
n = 1000
data = {
    'follower_count': np.random.randint(0, 5000, n),
    'following_count': np.random.randint(0, 5000, n),
    'post_count': np.random.randint(0, 100, n),
    'has_profile_pic': np.random.randint(0, 2, n) # 0: Yok, 1: Var
}
df = pd.DataFrame(data)

# Mantık: Takip ettiği kişi, takipçisinden 3 kat fazlaysa ve resmi yoksa %95 FAKE
df['is_fake'] = ((df['following_count'] > df['follower_count'] * 3) & (df['has_profile_pic'] == 0)).astype(int)

# 2. Model Eğitimi
X = df.drop('is_fake', axis=1)
y = df['is_fake']

model = RandomForestClassifier(n_estimators=100)
model.fit(X, y)

# 3. Modeli Kaydet
joblib.dump(model, 'fake_account_model.pkl')
print("✅ Classification Projesi 3: Fake Account Modeli Hazır!")
```
