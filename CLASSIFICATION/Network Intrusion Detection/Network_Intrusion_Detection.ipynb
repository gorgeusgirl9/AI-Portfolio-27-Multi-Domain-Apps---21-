```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import joblib

# 1. Sentetik Ağ Trafiği Verisi (Paket Boyutu ve Paket Aralığı)
np.random.seed(42)
n_traffic = 1000

# Paket Boyutu (KB) ve Paketler Arası Süre (ms)
# Normal trafik: Küçük paketler, düzenli aralıklar
# Saldırı (DDoS): Çok yoğun paketler veya çok büyük paketler
packet_size = np.random.uniform(1, 1500, n_traffic)
packet_interval = np.random.uniform(0.1, 100, n_traffic)

X = np.column_stack((packet_size, packet_interval))

# 2. K-Means Modeli (3 Küme: Güvenli, Şüpheli, Potansiyel Saldırı)
model = KMeans(n_clusters=3, random_state=42, n_init=10)
model.fit(X)

# 3. Modeli Kaydet
joblib.dump(model, 'network_security_model.pkl')
print("✅ Siber Güvenlik Kümeleme Modeli Hazır!")
```
