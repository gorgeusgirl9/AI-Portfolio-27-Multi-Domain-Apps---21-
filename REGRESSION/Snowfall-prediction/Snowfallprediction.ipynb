```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import joblib

# HATASIZ ÇÖZÜM: Kendi verimizi oluşturuyoruz (Linke ihtiyaç yok)
np.random.seed(42)
n_rows = 1000

data = {
    'Temperature': np.random.uniform(-15, 15, n_rows),
    'Humidity': np.random.uniform(20, 100, n_rows),
    'Wind_Speed': np.random.uniform(0, 50, n_rows),
    'Pressure': np.random.uniform(950, 1050, n_rows)
}

df = pd.DataFrame(data)

# Kar Yağışı Mantığı: Sıcaklık < 2 ve Nem > 60 ise kar yağsın (1), değilse (0)
df['Snowing'] = ((df['Temperature'] < 2) & (df['Humidity'] > 60)).astype(int)

# Model Eğitimi
X = df[['Temperature', 'Humidity', 'Wind_Speed', 'Pressure']]
y = df['Snowing']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Modeli Kaydet
joblib.dump(model, 'snow_model.pkl')
print("✅ 4. Proje Modeli (Kar Tahmini) BAŞARIYLA oluşturuldu: snow_model.pkl")
```
