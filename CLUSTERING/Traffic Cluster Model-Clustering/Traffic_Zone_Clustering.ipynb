```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import joblib

# 1. Sentetik Trafik Verisi (Araç Sayısı ve Ortalama Hız)
np.random.seed(42)
n_points = 600
vehicle_count = np.random.randint(50, 2000, n_points) # Saatlik araç
avg_speed = np.random.randint(10, 120, n_points)      # Km/s

X = np.column_stack((vehicle_count, avg_speed))

# 2. K-Means Modeli (Trafik bölgelerini 3 ana gruba ayırıyoruz)
# Akıcı, Yoğun, Kilitlenmiş
model = KMeans(n_clusters=3, random_state=42)
model.fit(X)

# 3. Modeli Kaydet
joblib.dump(model, 'traffic_cluster_model.pkl')
print("✅ Clustering Projesi 1: Trafik Bölgeleri Analizi Hazır!")
```
