```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import joblib

# 1. Sentetik Ekonomik Veri Oluşturma
np.random.seed(42)
n_rows = 1500

gold = np.random.uniform(1800, 2600, n_rows)   # Altın Ons ($)
silver = np.random.uniform(20, 35, n_rows)     # Gümüş Ons ($)
oil = np.random.uniform(60, 110, n_rows)       # Petrol Varil ($)

# Borsa Endeksi Formülü (Altın ve Petrol borsa endeksini pozitif etkiler gibi simüle ettik)
market_index = (gold * 1.5) + (silver * 12) + (oil * 20) + np.random.normal(0, 100, n_rows)

df = pd.DataFrame({'Gold': gold, 'Silver': silver, 'Oil': oil, 'Market_Index': market_index})

# 2. Model Eğitimi
X = df[['Gold', 'Silver', 'Oil']]
y = df['Market_Index']

model = LinearRegression()
model.fit(X, y)

# 3. Modeli Kaydet
joblib.dump(model, 'market_model.pkl')
print("✅ Proje 3: Global Market Modeli Başarıyla Kaydedildi!")
```
