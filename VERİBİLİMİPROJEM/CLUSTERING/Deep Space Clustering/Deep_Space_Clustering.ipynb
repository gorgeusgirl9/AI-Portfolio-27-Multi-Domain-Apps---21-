```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import joblib

# 1. Uzay Verisi Simülasyonu (800 gök cismi)
np.random.seed(42)
n_objects = 800

# Parlaklık (0-100), Uzaklık (100-50.000 Işık Yılı), Radyasyon (0.1-15.0 THz)
data = {
    'brightness': np.random.uniform(0, 100, n_objects),
    'distance': np.random.uniform(100, 50000, n_objects),
    'radiation': np.random.uniform(0.1, 15.0, n_objects)
}
df = pd.DataFrame(data)

# 2. K-Means Modeli (3 Küme)
# n_init=10: Algoritmanın daha kararlı çalışmasını sağlar.
model = KMeans(n_clusters=3, random_state=42, n_init=10)
model.fit(df)

# 3. Modeli Kaydet
joblib.dump(model, 'space_object_model.pkl')
print("✅ Uzay Nesnesi Kümeleme Modeli Başarıyla Kaydedildi!")
```
