```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import joblib

# 1. Eğitim Verisi (Basit bir örnek set)
data = {
    'text': [
        'bu ürün harika bayıldım', 'kesinlikle tavsiye ederim', 'çok güzel bir gün',
        'berbat bir ürün sakın almayın', 'hiç beğenmedim çok kötü', 'hayal kırıklığı'
    ],
    'sentiment': [1, 1, 1, 0, 0, 0] # 1: Pozitif, 0: Negatif
}
df = pd.DataFrame(data)

# 2. Metinleri Sayılara Çevirme (Vektörleştirme)
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['text'])

# 3. Model Eğitimi (Naive Bayes - NLP için klasiktir)
model = MultinomialNB()
model.fit(X, df['sentiment'])

# 4. Modelleri Kaydet (İki dosya lazım!)
joblib.dump(model, 'sentiment_model.pkl')
joblib.dump(vectorizer, 'vectorizer.pkl')

print("✅ NLP Projesi 1: Duygu Analizi Modeli Hazır!")
```
